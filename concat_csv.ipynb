```python
import pandas as pd
```

```python
jobkorea = pd.read_csv('./jobkorea/jobkorea_jobs.csv')
jumpit = pd.read_csv('./jumpit/jumpit_jobs.csv')
saramin = pd.read_csv('./saramin/saramin_jobs.csv')
wanted = pd.read_csv('./wanted/wanted_jobs_categorized_with_location.csv')
```

```python
jobkorea.head()
```

```python
jobkorea = jobkorea.drop(['typ'], axis=1)
jobkorea.head()
```

```python
jumpit.head()
```

```python
jumpit = jumpit[['company', 'title', 'url', 'exp', 'loc', 'job_name']]
jumpit.head()
```

```python
saramin.head()
```

```python
saramin = saramin.drop(['계약형태', '지역(구)'], axis=1)
saramin.head()
```

```python
saramin['지역'].unique()
```

```python
out = ['미국', '일본', '중국·홍콩', '아시아·중동', '북·중미', '남미', '유럽', '오세아니아', '아프리카', '남극대륙', '기타해외']

saramin['지역'] = saramin['지역'].map(lambda x: '해외' if x in out else x.replace('전체', '').replace('세종특별자치시', '세종'))
saramin.head()
```

```python
saramin['지역'].unique()
```

```python
wanted.head()
```

```python
wanted['지역'] = wanted['지역'].map(lambda x: x.split(' ')[0])
wanted.head()
```

```python
wanted['지역'].unique()
```

```python
wanted = wanted.drop(['location'], axis=1)
wanted.head()
```

```python
changeco = ['company', 'title', 'url', '경력', '지역', 'role']
wanted = wanted[changeco]
wanted.head()
```

```python
wanted = wanted[wanted['role'] != '기타']
wanted.head()
```

```python
one = ['회사이름', '공고이름', 'url', '경력', '지역', '직무']

jobkorea.columns = one
jumpit.columns = one
saramin.columns = one
wanted.columns = one
```

```python
jobkorea.head()
```

```python
jumpit.head()
```

```python
saramin.head()
```

```python
wanted.head()
```

```python
final = pd.concat([jobkorea, jumpit, saramin, wanted], ignore_index=True)
final.head()
```

```python
len(final)
```

```python
final[['회사이름', '공고이름', '직무']].duplicated().sum()
```

```python
final = final.drop_duplicates(subset=['회사이름', '공고이름', '직무']).reset_index(drop=True)
len(final)
```

```python
final[['회사이름', '공고이름', '직무']].duplicated().sum()
```

```python
final.to_csv('final.csv', index=False)
```
