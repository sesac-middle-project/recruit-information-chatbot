```python
import pandas as pd
```

```python
jobkorea = pd.read_csv('./jobkorea/jobkorea_jobs.csv')
jumpit = pd.read_csv('./jumpit/jumpit_jobs.csv')
saramin = pd.read_csv('./saramin/saramin_jobs.csv')
wanted = pd.read_csv('./wanted/wanted_jobs_categorized_with_location.csv')
```

```python
jobkorea.head()
```

```python
jobkorea = jobkorea.drop(['typ'], axis=1)
jobkorea.head()
```

```python
jumpit.head()
```

```python
jumpit = jumpit[['company', 'title', 'url', 'exp', 'loc', 'job_name']]
jumpit.head()
```

```python
saramin.head()
```

```python
saramin = saramin.drop(['계약형태', '지역(구)'], axis=1)
saramin.head()
```

```python
saramin['지역'].unique()
```

```python
out = ['미국', '일본', '중국·홍콩', '아시아·중동', '북·중미', '남미', '유럽', '오세아니아', '아프리카', '남극대륙', '기타해외']

saramin['지역'] = saramin['지역'].map(lambda x: '해외' if x in out else x.replace('전체', '').replace('세종특별자치시', '세종'))
saramin.head()
```

```python
saramin['지역'].unique()
```

```python
wanted.head()
```

```python
wanted['지역'] = wanted['지역'].map(lambda x: x.split(' ')[0])
wanted.head()
```

```python
wanted['지역'].unique()
```

```python
wanted = wanted.drop(['location'], axis=1)
wanted.head()
```

```python
changeco = ['company', 'title', 'url', '경력', '지역', 'role']
wanted = wanted[changeco]
wanted.head()
```

```python
wanted = wanted[wanted['role'] != '기타']
wanted.head()
```

```python
one = ['회사이름', '공고이름', 'url', '경력', '지역', '직무']

jobkorea.columns = one
jumpit.columns = one
saramin.columns = one
wanted.columns = one
```

```python
jobkorea.head()
```

```python
jobkorea['사이트'] = 'jobkorea'
jobkorea.head()
```

```python
jumpit.head()
```

```python
jumpit['사이트'] = 'jumpit'
```

```python
saramin.head()
```

```python
saramin['사이트'] = 'saramin'
```

```python
wanted.head()
```

```python
wanted['사이트'] = 'wanted'
```

```python
final = pd.concat([jobkorea, jumpit, saramin, wanted], ignore_index=True)
final.head()
```

```python
len(final)
```

```python
final['회사이름'] = final['회사이름'].str.replace(r"\(주\)(?: 이모티콘)?|㈜|\s+|\(.*?\)", "", regex=True)
```

```python
final[['회사이름', '공고이름']].duplicated().sum()
```

```python
duplicated_rows = final[final[['회사이름', '공고이름']].duplicated(keep=False)]
grouped_counts = duplicated_rows.groupby(['회사이름', '공고이름']).size().reset_index(name='중복횟수')

print(grouped_counts)
```

```python
final = final.drop_duplicates(subset=['회사이름', '공고이름']).reset_index(drop=True)
len(final)
```

```python
final[['회사이름', '공고이름']].duplicated().sum()
```

```python
final.to_csv('final.csv', index=False)
```

```python
da = final[final['직무'] == '데이터 분석가']
de = final[final['직무'] == '데이터 엔지니어']
ai = final[final['직무'] == 'AI 개발자']
cb = final[final['직무'] == '챗봇 개발자']
ce = final[final['직무'] == '클라우드 엔지니어']
api = final[final['직무'] == 'API 개발자']
me = final[final['직무'] == '머신러닝 엔지니어']
ds = final[final['직무'] == '데이터 사이언티스트']
```

```python
da.to_csv('da.csv', index=False)
de.to_csv('de.csv', index=False)
ai.to_csv('ai.csv', index=False)
cb.to_csv('cb.csv', index=False)
ce.to_csv('ce.csv', index=False)
api.to_csv('api.csv', index=False)
me.to_csv('me.csv', index=False)
ds.to_csv('ds.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = 'False'
```

```python
# 신입과 경력 포함 여부를 분리
newbie = final[final['경력'].str.contains('신입', na=False)]
experienced = final[final['경력'].str.contains('경력', na=False)] 

# 신입과 경력이 둘 다 포함된 직무
both = newbie[newbie['경력'].str.contains('경력', na=False)]

# 신입만 포함된 직무 (신입이 포함되고 경력은 포함되지 않음)
only_newbie = newbie[~newbie['경력'].str.contains('경력', na=False)]

# 경력만 포함된 직무 (경력이 포함되고 신입은 포함되지 않음)
only_experienced = experienced[~experienced['경력'].str.contains('신입', na=False)]

# 각 카운트
counts = {
    '신입 & 경력': both.shape[0],
    '신입': only_newbie.shape[0],
    '경력': only_experienced.shape[0]
}

# 시각화
plt.figure(figsize=(6, 4))
plt.bar(counts.keys(), counts.values(), color=['blue', 'green', 'orange'])
plt.title('Job Title Categorization: 신입 vs 경력')
plt.xlabel('Experience')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 신입, 경력, 신입&경력 카테고리 생성
graph = final.copy()

graph['경력여부'] = graph['경력'].apply(
    lambda x: '신입 & 경력' if '신입' in str(x) and '경력' in str(x) else
              ('신입' if '신입' in str(x) else '경력' if '경력' in str(x) else '기타')
)

# 직무별로 신입, 경력, 신입&경력 카운트
exp_dist = graph.groupby(['직무', '경력여부']).size().unstack().fillna(0)

# 시각화
plt.figure(figsize=(10, 5))
exp_dist.plot(kind='bar', stacked=True, ax=plt.gca(), color=['blue', 'green', 'orange'])
plt.title('Experience Distribution by Job Name')
plt.xlabel('Job Name')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
