```python
# Define mappings for more specific roles based on `role` or `title` keywords
role_mapping = {
    "머신러닝 엔지니어": "머신러닝 엔지니어",
    "데이터 엔지니어": "데이터 엔지니어",
    "데이터 사이언티스트": "데이터 사이언티스트",
    "AI 서비스 개발자": ["AI", "Artificial Intelligence"],
    "챗봇 개발자": ["챗봇", "Chatbot"],
    "클라우드 엔지니어": ["클라우드", "Cloud"],
    "API 개발자": ["API"]
}

# Function to categorize based on title keywords
def categorize_job(row):
    for new_role, keywords in role_mapping.items():
        if isinstance(keywords, list):
            if any(keyword in row['title'] for keyword in keywords):
                return new_role
        elif keywords == row['role']:
            return new_role
    return "기타"

# Apply categorization to the dataset
jobs_data['specific_role'] = jobs_data.apply(categorize_job, axis=1)

# Save the categorized data for user
categorized_file_path = '/mnt/data/wanted_jobs_categorized.csv'
jobs_data.to_csv(categorized_file_path, index=False)

# Display the results
import ace_tools as tools; tools.display_dataframe_to_user(name="Categorized Job Listings", dataframe=jobs_data)

categorized_file_path
```
