```python
job_type = ['정규직', '인턴직', '계약직', '정규직·계약직', '정규직·인턴직',
       '정규직·기간제', '기간제·계약직', '정규직·프리랜서', '계약직·인턴직', '정규직·해외취업',
        '파트·프리랜서', '무기계약직·계약직', '프리랜서', '계약직·프리랜서', '정규직·무기계약직']
```

```python
df_type = df[df['계약형태'].str.contains('|'.join(job_type), na=False)]
df_type[20:25]
```

```python
df_lower = df_type.copy()

df_lower['제목'] = df_lower['제목'].str.lower()
```

```python
def check_keywords(row, keywords):
    for keyword_group in keywords:
        if all(keyword in row for keyword in keyword_group):
            return True
    return False
```

```python
da_with = [
    ['데이터', '분석'],
    ['데이터', '분석가'],     
    ['data', 'analyst'],  
    ['db', '분석'],      
    ['데이터', '애널리스트'],
    ['buisness', 'analyst'],
    ['analyst'],   
    ['분석'],     
    ['데이터 분석'], 
    ['data analyst'] 
]
```

```python
filt = df_lower[df_lower['직무'] == '데이터 분석가']

da = filt[filt['제목'].apply(lambda x: check_keywords(x, da_with))]
da.head()
```

```python
de_with = [
    ['데이터', '엔지니어'],    
    ['data', 'engineer'],  
    ['db', '엔지니어'],      
    ['db', 'engineer']
]
```

```python
filt = df_lower[df_lower['직무'] == '데이터 엔지니어']

de = filt[filt['제목'].apply(lambda x: check_keywords(x, de_with))]
de.head()
```

```python
ai_with = [
    ['a', 'i', '개발'],
    ['인공지능', '개발'],
    ['a', 'i', '엔지니어'],
    ['a', 'i', 'engineer'],
    ['a', 'i', '연구'],
    ['a', 'i', 'architect'],
    ['딥', '러닝'],
    ['deep', 'learning'],
    ['machine', 'learning'],
    ['a', 'i', 'researcher'],
    ['a', 'i', 'solution'], 
    ['a', 'i', '전문가']
]
```

```python
filt = df_lower[df_lower['직무'] == 'AI 개발자']

ai = filt[filt['제목'].apply(lambda x: check_keywords(x, ai_with))]
ai.head()
```
